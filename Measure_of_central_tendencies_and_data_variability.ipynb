```python
## importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # Visualization
%matplotlib inline
```

```python
## generating a random variable
exp=np.random.normal(25000,15000,1000) #mean,standard deviation, number of datapoints
```

```python
exp
```

```python
## plotting the randomly generated exp
pd.DataFrame(exp).plot(kind='hist')
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(exp)
```

```python
## calculating mean before adding outlier
np.mean(exp)
```

```python
## calculating median before adding outlier
np.median(exp)
```

```python
## adding one huge point,outlier
exp=np.append(exp,[100000000])
```

```python
sns.kdeplot(exp)
```

```python
#Mean after outlier
np.mean(exp)
```

```python
#Median after outlier
np.median(exp)
```

```python
## mode
exp=np.random.randint(15,high=50,size=200) # Generates 200 random integers between 15-50
```

```python
sns.histplot(exp)
```

```python
from scipy import stats
stats.mode(exp)
```
